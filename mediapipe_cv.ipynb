```python
import cv2
import mediapipe as mp


mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

with mp_pose.Pose(
    static_image_mode=True) as pose:   #PARA VIDEOS PONER FALSE /TRUE ES PARA FOTOS
    image = cv2.imread("data/dataset_train/IMG_5694_032.jpg")
    image = cv2.resize(src = image, dsize = (771,1370))
    height, width , _ = image.shape
    image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

    results = pose.process(image_rgb)
    print ("Pose landmarks:", results.pose_landmarks)

    if results.pose_landmarks is not None:
        mp_drawing.draw_landmarks(image, results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            mp_drawing.DrawingSpec(color=(43,214,38), thickness = 2, circle_radius= 3),   #personalizar el esqueleto
            mp_drawing.DrawingSpec(color = (255,255,255), thickness = 2))

    cv2.imshow("Image", image)
    cv2.waitKey(0)
cv2.destroyAllWindows()



```

```python
#con video 

import cv2
import mediapipe as mp


mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

cap = cv2.VideoCapture("data/dataset_train/IMG_5694_032.jpg")


with mp_pose.Pose(
    static_image_mode=False) as pose:   #PARA VIDEOS PONER FALSE /TRUE ES PARA FOTOS
        while True:
             ret, frame = cap.read()
             if ret == False:
                  break
             heigth, width, _ = frame.shape 
             frame_rgb= cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
             results = pose.process(frame_rgb)

             if results.pose_landmarks is not None:
               mp_drawing.draw_landmarks(
                    frame, results.pose_landmarks,
                    mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(43,214,38), thickness = 2, circle_radius= 3),   #personalizar el esqueleto
                    mp_drawing.DrawingSpec(color = (255,255,255), thickness = 2))

          
             
             cv2.imshow("Frame", frame)
             if cv2.waitKey(1) & 0xFF == 27:
                  break
             
cap. realase()
cv2.destroyAllWindows()
    

    
```

```python
#con video streaming

import cv2
import mediapipe as mp


mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)


with mp_pose.Pose(
    static_image_mode=False) as pose:   #PARA VIDEOS PONER FALSE /TRUE ES PARA FOTOS
        while True:
             ret, frame = cap.read()
             if ret == False:
                  break
             frame = cv2.flip(frame,1)  #espejo a la camara 
             heigth, width, _ = frame.shape 
             frame_rgb= cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
             results = pose.process(frame_rgb)

             if results.pose_landmarks is not None:
               mp_drawing.draw_landmarks(
                    frame, results.pose_landmarks,
                    mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(43,214,38), thickness = 2, circle_radius= 3),   #personalizar el esqueleto
                    mp_drawing.DrawingSpec(color = (255,255,255), thickness = 2))

          
             
             cv2.imshow("Frame", frame)
             if cv2.waitKey(1) & 0xFF == 27:
                  break
             
cap. realase()
cv2.destroyAllWindows()
```
